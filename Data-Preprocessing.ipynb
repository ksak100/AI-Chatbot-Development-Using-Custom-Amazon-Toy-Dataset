```python
data=pd.read_csv("C:/Users/kunal/Downloads/Book11.csv")
import pandas as pd

# Splitting 'QA' column into 'Question' and 'Answer' columns
data['Question'] = data['QA'].apply(lambda x: x.split('Question')[-1] if 'Question' in x else None)
data['Answer'] = data['QA'].apply(lambda x: x.split('Answer')[-1] if 'Answer' in x else None)

# Dropping rows where 'Question' or 'Answer' is None
#data = data.dropna(subset=['Question', 'Answer'])

# Dropping the original 'QA' column
data = data.drop(columns=['QA'])

data
```

```python
data['Answer'] = data['Answer'].shift(-1)
```

```python
data.dropna(subset=['Question', 'Answer'], how='all', inplace=True) 
```
